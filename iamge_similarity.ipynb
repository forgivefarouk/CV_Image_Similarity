```python
from keras.applications.resnet50 import ResNet50, preprocess_input
import tensorflow as tf
import numpy as np
import os
import cv2
import pandas as pd
import pathlib
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
  model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
  model.trainable = False
```

```python
model= tf.keras.Sequential([
  model,
  tf.keras.layers.GlobalAveragePooling2D(),  
])
```

```python
model.summary()
```

```python
def feature_extractor(image_path,model):
  try:
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)
    results = model.predict(img).flatten()
    normalized_results = results / np.linalg.norm(results)
    return normalized_results
  except:
    print("An error occurred while extracting features for image:", image_path)
    return None
```

```python
path = pathlib.Path('dataset')
```

```python
feature_list =[]
file_name = []
for img in tqdm(os.listdir(path)):
  results = feature_extractor(f'{path}/{img}',model)
  if results is not None:
    feature_list.append(results)
    file_name.append(os.path.join(path,img))
```

```python
import pickle
""" pickle.dump(feature_list,open('feature_list.pkl','wb'))
pickle.dump(file_name,open('file_name.pkl','wb')) """
features = np.array(pickle.load(open('feature_list.pkl','rb')))
file_name = pickle.load(open('file_name.pkl','rb'))
```

```python
from sklearn.neighbors import NearestNeighbors

nbrs = NearestNeighbors(n_neighbors=5, algorithm='brute', metric='euclidean')
nbrs.fit(features)
```

```python
def recommendations(features,top_n,image,model):
    nbrs = NearestNeighbors(n_neighbors=top_n, algorithm='brute', metric='euclidean')
    nbrs.fit(features)
    normalized_results=feature_extractor(image,model)
    dis , idx = nbrs.kneighbors([normalized_results])
    return idx
    
```
